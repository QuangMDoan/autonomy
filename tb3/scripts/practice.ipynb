```python
import numpy as np 

a = np.ones((4, 6))    # a is a (4, 6) 2d array
b = np.ones((4, 1))    # b is a (4, 1) 2d array
c = np.ones((1, 6))     # c is a (1, 6) 2d array
d = np.ones(6)          # d is a size-6 1d array

print(a.shape, "\n", a)
```

```python
x = a[:, 0]
print(f"a[:, 0] = {x}, shape: {x.shape}")

print()


y = a[0, :]
print(f"a[0, :] = {y}, shape: {y.shape}")
```

```python
print(f"a, shape: {a.shape} \n {a}")
print()
x = a[:, 0:3]
print(f"a[:, 0:3], shape: {x.shape} \n {x}")
```

```python
print(f"a, shape: {a.shape} \n {a}")
print()
x = a[None, :, :]
print(f"a[None, :, :], shape: {x.shape} \n {x}")
```

```python
print(f"a, shape: {a.shape} \n {a}")
print()

# x.shape = (4, 1, 6)
x = a[:, None, :]
print(f"a[:, None, :], shape: {x.shape} \n {x}")
```

```python
a = np.ones((4, 6)) 
b = 2 * np.ones((4, 1)) 

print(f"a, shape: {a.shape} \n {a}")
print(f"b, shape: {b.shape} \n {b}")

print(f"a+b, shape: {(a+b).shape} \n {a+b}")
```

```python
a = np.ones((1, 6)) 
b = 2 * np.ones((4, 1)) 

print(f"a, shape: {a.shape} \n {a}")
print(f"b, shape: {b.shape} \n {b}")

print(f"a+b, shape: {(a+b).shape} \n {a+b}")
```

```python
a = np.ones(6)
b = 2 * np.ones((4, 1)) 

print(f"a, shape: {a.shape} \n {a}")
print(f"b, shape: {b.shape} \n {b}")

print(f"a+b, shape: {(a+b).shape} \n {a+b}")
```
